```python
potreb1=pd.read_excel('./data/температура1.xlsx',skiprows=1,sheet_name=1,header=[0])
potreb1 = potreb1.fillna(method='ffill', axis=0)
potreb1['data_g']=potreb1[ 'Год'].astype(int)
potreb1['data_m']=potreb1['Месяц'].astype(int)
potreb1['data_d']=potreb1[ 'День'].astype(int)
potreb1['data']=pd.to_datetime(potreb1['data_g'].astype(str)+'-'+potreb1['data_m'].astype(str) +'-' + potreb1['data_d'].astype(str))
potreb1['data_temp']=potreb1['data']
potreb1=potreb1.set_index('data')
potreb1=potreb1.drop(([ 'data_g','data_m','data_d','data_temp', 'Год','Месяц','День']), axis=1)/1000

potreb1
```

```python
potreb2=pd.read_excel('./data/температура2.xlsx',skiprows=1,sheet_name=1,header=[0])
potreb2 = potreb2.fillna(method='ffill', axis=0)
potreb2['data_g']=potreb2[ 'Год'].astype(int)
potreb2['data_m']=potreb2['Месяц'].astype(int)
potreb2['data_d']=potreb2[ 'День'].astype(int)
potreb2['data']=pd.to_datetime(potreb2['data_g'].astype(str)+'-'+potreb2['data_m'].astype(str) +'-' + potreb2['data_d'].astype(str))
potreb2['data_temp']=potreb2['data']
potreb2=potreb2.set_index('data')
potreb2=potreb2.drop(([ 'data_g','data_m','data_d','data_temp', 'Год','Месяц','День']), axis=1)/1000

```

```python
potreb=pd.concat([potreb1, potreb2], axis=1)
```

```python
#https://data.gov.ru/opendata/7710349494-prognozkons/data-20190213T1046-structure-20190213T1046.csv

ml=pd.read_csv('https://data.gov.ru/opendata/7710349494-prognozkons/data-20190213T1046-structure-20190213T1046.csv',index_col=0).T
ml
```

```python

ml=ml.iloc[1:]
ml=ml.apply(pd.to_numeric, errors='coerce')
ml.index=ml.index.astype('int32')
```

```python
for i in ml.columns:
    print('"',i,'"')
```

```python
potreb_god=potreb.groupby(by=[potreb.index.year], dropna=False).sum()
potreb_god.iloc[-1]=12*potreb_god.iloc[-1]/11
potreb_god
```

```python
cols=['Производительность труда','Индекс  потребительских цен  на конец года','Темп роста в реальном выражении','Общая численность безработных граждан','Валовый внутренний продукт Номинальный объем','Курс Доллара','Оборот розничной торговли Номинальный объем']
```

```python
ml[cols]
```

```python
ml[cols].loc[potreb_god.index].fillna(0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, median_absolute_error, r2_score

kof=pd.DataFrame()
model={}
pred=pd.DataFrame(index=ml.index)
for reg in potreb_god:
    model[reg] = LinearRegression()
    model[reg].fit(ml[cols].loc[potreb_god.index][:-1].fillna(0), potreb_god[reg][:-1])
    kof=kof.append({'reg':reg, 'A':model[reg].coef_.tolist()[0], 'B':model[reg].coef_.tolist()[1],'C':model[reg].intercept_}, ignore_index=True)
    pred[reg]=model[reg].predict(ml[cols].fillna(0))
```

```python
(1-((potreb_god-pred)/potreb_god).loc[2020].abs().mean())*100
```
